# Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

A1. 

The Pearson correlation coefficient, denoted as *r*, is a measure of the linear relationship between two continuous variables. It quantifies the strength and direction of the linear association between the variables. The value of *r* ranges from -1 to 1, with the following interpretations:

*r* = 1: Perfect positive linear relationship. As one variable increases, the other increases linearly.

*r* = −1: Perfect negative linear relationship. As one variable increases, the other decreases linearly.

*r* = 0: No linear relationship. The variables are not linearly related.

If *r* is close to 1 (positive value close to 1): There is a strong positive linear relationship between the amount of time spent studying and exam scores. As students spend more time studying, their exam scores tend to increase linearly.

If *r* is close to -1 (negative value close to -1): There is a strong negative linear relationship between the amount of time spent studying and exam scores. As students spend more time studying, their exam scores tend to decrease linearly.

If *r* is close to 0: There is little to no linear relationship between the amount of time spent studying and exam scores. Changes in study time do not linearly predict changes in exam scores.

Interpreting the Pearson correlation coefficient should also consider the magnitude of *r*. A larger absolute value of *r* indicates a stronger linear relationship, while values closer to 0 indicate weaker or no linear relationship. Additionally, it's important to remember that correlation does not imply causation. Even if a strong correlation is found, it does not necessarily mean that studying more causes higher exam scores; other factors may be involved.

To calculate *r*, follow these steps:

1. Calculate the mean (ˉXˉ) of the number of hours individuals spend watching television per day.
2. Calculate the mean (ˉYˉ) of the level of physical activity.
3. Calculate the differences between each data point and its respective mean for both variables (i.e., (Xi − ˉXˉ) and (Yi - ˉYˉ)).
